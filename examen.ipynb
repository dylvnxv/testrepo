import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

tables = soup.find_all('table')
revenue_table = tables[1]

rows = revenue_table.find_all('tr')


data = []
for row in rows[1:]:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)


tesla_revenue = pd.DataFrame(data, columns=['Date', 'Revenue'])

tesla_revenue['Revenue'] = tesla_revenue['Revenue'].str.replace('$', '').str.replace(',', '')

print(tesla_revenue.tail())

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to make a graph
def make_graph(stock_data, stock_name):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Date'], stock_data['Close'], label=f'{stock_name} Close Price')
    plt.title(f'{stock_name} Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')

# Reset the index to have 'Date' as a column
tesla_data_reset = tesla_data.reset_index()

# Call the make_graph function to plot the Tesla stock data
make_graph(tesla_data_reset, 'Tesla')
```

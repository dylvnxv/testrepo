```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
from io import StringIO

url = 'https://finance.yahoo.com/quote/TSLA/financials'
r = requests.get(url)
s = BeautifulSoup(r.content, 'html.parser')
t = s.find_all('table')
tesla_revenue_table = t[0]
html_data = StringIO(str(tesla_revenue_table))
tesla_revenue = pd.read_html(html_data)[0]
print(tesla_revenue.tail())
tesla_revenue.to_csv('tesla_revenue.csv', index=False)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")
tesla_data_reset = tesla_data.reset_index()

make_graph(tesla_data_reset, 'Tesla Stock Price Over the Last 5 Years')
```
